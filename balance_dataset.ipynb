```python
# import dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
print("Loading test data from files...")
test_data = np.load('/share/pi/hackhack/Breast/Breast_MRI/ericjason230/cleaned_data_test.npy', allow_pickle=True)
test_labels = np.load('/share/pi/hackhack/Breast/Breast_MRI/ericjason230/cleaned_label_test.npy', allow_pickle=True)
print("Test data loaded")
```

```python
plt.figure(figsize=[5,5])

# Display the first image in training data
plt.subplot(121)
curr_img = np.reshape(test_data[0], (512,512))
plt.imshow(curr_img, cmap='gray')

# Display the first image in testing data
plt.subplot(122)
curr_img = np.reshape(test_labels[0], (512,512))
plt.imshow(curr_img, cmap='gray')
```

```python
zero_count = 0
nonzero_count = 0
final_indexes = []
for i, test_label in enumerate(test_labels):
    if np.sum(test_label) == 0.0:
        zero_count += 1
    else:
        final_indexes.append(i)
        nonzero_count += 1
print(final_indexes)
```

```python
final_test_data = np.take(test_data, final_indexes, axis=0)
final_test_labels = np.take(test_labels, final_indexes, axis=0)
```
